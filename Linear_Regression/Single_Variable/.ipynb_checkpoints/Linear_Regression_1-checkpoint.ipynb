```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn import linear_model
import joblib
import pickle
```

```python
df = pd.read_csv("homeprices.csv")
df
```

```python
%matplotlib inline
plt.xlabel("Area")
plt.ylabel("Price")
plt.scatter(df.area, df.price)
```

```python
lr = linear_model.LinearRegression()
lr.fit(df[["area"]] ,df.price)
```

```python
lr.predict([[2600]])
```

```python
with open ('model_pickle', 'wb') as f:
    pickle.dump(lr,f)
```

```python
with open ('model_pickle', 'rb') as f:
    model_p = pickle.load(f)
```

```python
model_p.predict([[2600]])
```

```python
joblib.dump(model_p, 'model_joblib')
```

```python
model_j = joblib.load('model_joblib')
```

```python
model_j.predict([[2600]])
```

```python
%matplotlib inline
plt.xlabel("Area")
plt.ylabel("Price")
plt.scatter(df.area, df.price)
plt.plot(df.area, lr.predict(df[["area"]]))
```

```python
df_new = pd.read_csv("excercise.csv")
df_new
```

```python
df_new = df_new.rename(columns = {"per capita income (US$)": "value"})
```

```python
%matplotlib inline
plt.xlabel("Year")
plt.ylabel("PCI (US$)")
plt.scatter(df_new.year, df_new.value)
```

```python
lr_new = linear_model.LinearRegression()
lr_new.fit(df_new[["year"]], df_new.value)
```

```python
lr_new.predict([[2014]])
```

```python
print(lr_new.coef_)
print(lr_new.intercept_)
print(828.46507522*2016+(-1632210.7578554575))
```

```python
%matplotlib inline
plt.xlabel("Year")
plt.ylabel("PCI (US$)")
plt.scatter(df_new.year, df_new.value)
plt.plot(df_new.year, lr_new.predict(df_new[["year"]]))
plt.autoscale()
```
